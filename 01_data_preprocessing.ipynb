# Data Preprocessing

This notebook performs basic data preprocessing for the financial decision-making datasets used in the project. 
It includes simulated data creation and feature scaling to prepare inputs for subsequent modelling and explainability analysis.

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
```

```python
df = pd.DataFrame({
    'age': [25, 40, 35, 50],
    'income': [30000, 60000, 45000, 80000],
    'credit_score': [620, 720, 680, 750],
    'default': [1, 0, 0, 0]
})
```

```python
scaler = StandardScaler()
df[['age', 'income', 'credit_score']] = scaler.fit_transform(
    df[['age', 'income', 'credit_score']]
)

df
```
